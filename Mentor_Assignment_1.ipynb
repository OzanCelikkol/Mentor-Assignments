**T-TEST**

**One Sample T-Test**

One sample T-Test is used to determine difference between a defined constant and mean of values of a variable.
<Br>
For example, consider we have a data set about passenger on the Titanic. We can use age values of the passengers for this test.
<Br>
First of all we must determine our hypothesises.
<Br>
M = 50
<Br>
H0 = Mean of the ages of the passengers is similar to 50 with 5% error.
<Br>
HA = Mean of the ages of the passengers is not similar to 50 with 5% error.

```python
import numpy as np
import pandas as pd

df = pd.read_csv(r"C:\Users\Kleerys\Desktop\Career\1.1-Python\1.1-DataSets\train.csv")
display(df)
dfAge = pd.DataFrame(df["Age"])
display(dfAge)
```

Now we have the passengers' ages as a dataframe.
<Br>
Now we will check if mean of these values is similar to 50 by one sample t-test.
    <Br>
        To do this we have to import scipy.stats.

```python
import scipy.stats as scis

mean = dfAge.Age.mean()
dfAge.fillna(mean, inplace = True)
print(scis.ttest_1samp(dfAge.Age, 50))
```

We filled NaN values with the mean of the age column then we do the test. As we can see p-value too small compares to 0.05.
<Br>
    According the results we should choose alternative hypothesis.

**Two Sample T-Test**
<Br>
Two sample T-Test is used to see difference between means of two sample populations.
    <Br>
        We will choose two sample populations that include randomly 500 values in our dataframe named as **dfAge**.

```python
sample1 = np.random.choice(dfAge.Age, 500)
sample2 = np.random.choice(dfAge.Age, 500)

print(scis.ttest_ind(sample1, sample2, equal_var = False, nan_policy = "raise"))
```

If we can accept error rate is the same, as we can see p-values is greater than 0.05. So we should stick on the zero <Br>hypothesis. **These test results give us a clue about our sample populations could be same.**
